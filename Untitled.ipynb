```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df.drop_duplicates()
```

```python
def quality_label(q):
    if q <= 4:
        return "low"
    elif q <= 6:
        return "medium"
    else:
        return "high"

df['quality_label'] = df['quality'].apply(quality_label)
```

```python
from sklearn.preprocessing import StandardScaler

X = df.drop(['quality', 'quality_label'], axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, clusters)
    print("k =", k, "Silhouette score =", score)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X_scaled)

sns.scatterplot(data=df, x='alcohol', y='pH', hue='cluster', palette='Set1')
plt.title("K-means clustering, k=3")
plt.show()
```

```python
pd.crosstab(df['quality_label'], df['cluster'])
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

Z = linkage(X_scaled, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(Z)
plt.title("Hierarchical clustering dendrogram")
plt.xlabel("Wine samples")
plt.ylabel("Distance")
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

for t in [10, 15, 20]:
    hier_clusters = fcluster(Z, t=t, criterion='distance')
    print("Cut-off distance:", t)
    print("Number of clusters:", len(set(hier_clusters)))
    print(pd.crosstab(df['quality_label'], hier_clusters))
    print()
```

```python
from sklearn.model_selection import train_test_split

y = df['quality_label']
X = df.drop(['quality', 'quality_label', 'cluster'], axis=1)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
y_test.value_counts()
```

```python
y_train.value_counts()
```

```python
print("Training size:", len(X_train))
print("Test size:", len(X_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

for c in [0.1, 1, 10]:
    lr = LogisticRegression(max_iter=1000, C=c)
    lr.fit(X_train, y_train)
    pred = lr.predict(X_test)
    print("C =", c, "Accuracy =", accuracy_score(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

for n in [50, 100, 200]:
    rf = RandomForestClassifier(n_estimators=n, random_state=42)
    rf.fit(X_train, y_train)
    pred = rf.predict(X_test)
    print("n_estimators =", n, "Accuracy =", accuracy_score(y_test, pred))
```

```python
from sklearn.neural_network import MLPClassifier

configs = [(50,), (100,), (50,50)]

for config in configs:
    nn = MLPClassifier(hidden_layer_sizes=config, max_iter=500)
    nn.fit(X_train, y_train)
    pred = nn.predict(X_test)
    print("Layers =", config, "Accuracy =", accuracy_score(y_test, pred))
```

```python
from sklearn.metrics import accuracy_score

# ANN (best model)
ann = MLPClassifier(hidden_layer_sizes=(50,), max_iter=500)
ann.fit(X_train, y_train)
pred_ann = ann.predict(X_test)

# Logistic Regression (best model)
lr = LogisticRegression(max_iter=1000, C=0.1)
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)

# Random Forest (best model)
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)

# Print results
print("ANN Accuracy:", accuracy_score(y_test, pred_ann))
print("Logistic Regression Accuracy:", accuracy_score(y_test, pred_lr))
print("Random Forest Accuracy:", accuracy_score(y_test, pred_rf))
```
